## 1.1. Sobre los Datos Abiertos


### Concepto

Datos Abiertos, conocidos en inglés como open data, es una iniciativa global vinculada a las políticas de Gobierno Abierto. Su objetivo es que los datos e información, especialmente aquellos generados por las administraciones públicas, se publiquen de manera abierta, regular y reutilizable por cualquier persona u organización. Esto implica que no haya restricciones de acceso, derechos de autor, patentes u otros mecanismos de control.

En Paraguay, los Datos Abiertos Gubernamentales son aquellos que las instituciones del estado ponen a disposición libremente para sus ciudadanos. Esta práctica busca brindar transparencia en la gestión pública y promover la rendición de cuentas. Además, fomenta la participación y colaboración entre la ciudadanía y el gobierno, facilitando la gobernanza y creando valor público. Esto no solo mejora los servicios gubernamentales y promueve la innovación, sino que también permite a los ciudadanos exigir cambios y mejoras.

El acceso a estos datos impulsa el crecimiento económico y abre nuevas oportunidades para proporcionar competitividad en igualdad de condiciones a todos los ciudadanos y sectores.

Los datos abiertos pueden ser utilizados de diversas formas:

- Para ser simplemente consultados.
- Para mejorar la información disponible añadiendo nuevos datos y generando valor público.
- Para desarrollar nuevas aplicaciones y servicios.
- Para generar nuevos negocios.

### Formatos de los Datos

Los datos catalogados en [www.datos.gov.py](http://www.datos.gov.py/) se presentan en varios formatos. Preferiblemente, se utilizan formatos no propietarios, que no requieren la compra de paquetes de software comercial para poder leerlos, procesarlos y analizarlos. Algunos de los formatos disponibles son:

- **CSV (Comma Separated Values):** Archivos separados por comas.
- **XML (eXtensible Markup Language):** Describe los datos de manera estructurada utilizando etiquetas no predefinidas, favoreciendo la interoperabilidad.
- **TXT (Plain Text):** Documentos de texto plano que pueden utilizar diferentes codificaciones de caracteres.
- **KML (Keyhole Markup Language):** Un lenguaje de marcado basado en XML para representar datos geográficos en tres dimensiones.
- **SHP (Shapefile):** Un formato propietario estándar de datos espaciales, desarrollado por ESRI, que almacena tanto geometría como información alfanumérica, aunque no está preparado para almacenar información topológica. Este formato es un estándar de facto para el intercambio de información geográfica entre Sistemas de Información Geográfica (SIG).
- **ZIP:** Un formato de compresión sin pérdida, comúnmente utilizado para la compresión de documentos, imágenes o programas.
- **RDF (Resource Description Framework):** No es un formato concreto sino una infraestructura para describir recursos de la red mediante expresiones sujeto-predicado-objeto. Es una tecnología esencial para la web semántica, ya que permite añadir significado a las páginas combinándola con otras herramientas.

## 1.2. Referencias Legales sobre Datos Abiertos en Paraguay


En Paraguay, el marco legal y regulatorio de los datos abiertos se apoya en una serie de leyes y decretos que garantizan el acceso a la información pública, promueven la transparencia y aseguran la protección de datos personales. A continuación, se describen las principales normativas que sustentan estas políticas.

### Constitución Nacional - Artículo 135

El artículo 135 de la Constitución Nacional de Paraguay establece que todas las personas tienen el derecho de acceder a la información y los datos que obren en registros oficiales de carácter público, especialmente aquellos que conciernen a ellos mismos o a sus bienes. Este derecho fundamental respalda la transparencia y la rendición de cuentas del gobierno, garantizando que los ciudadanos puedan obtener información relevante y precisa sobre asuntos que les afectan directamente.

### Licencia de Uso de la Información Pública - Decreto 4064 - Ley Nro. 5282/2014

Esta licencia permite el uso gratuito, perpetuo y no exclusivo de la información y datos abiertos públicos del Estado Paraguayo a cualquier persona física o jurídica. Las actividades autorizadas incluyen copia, reproducción, distribución, comunicación pública, adaptación y transformación, bajo cualquier formato y modalidad permitida por la ley.

**Condiciones de Uso**

Para mantener la validez de la licencia, es necesario cumplir con estas condiciones:

- Citar la fuente pública que proporcionó la información o datos.
- Mencionar la fecha de la última actualización de la información o datos utilizados, si es conocida.
- No simular un uso oficial o patrocinado por el Estado Paraguayo con la información o datos.

**Exclusión de Responsabilidad**

El uso de la información y datos bajo esta licencia es responsabilidad exclusiva del usuario. La fuente proveedora no se hace responsable por cualquier daño derivado del uso o transformación de los datos por parte del usuario.



### Ley N° 6.207/2018 que Crea el Ministerio de Tecnologías de la Información y Comunicación (MITIC)

La Ley N° 6.207, promulgada en 2018, establece la creación del Ministerio de Tecnologías de la Información y Comunicación (MITIC). Este ministerio es responsable de formular y coordinar las políticas públicas en materia de tecnologías de la información, incluyendo la promoción de la transparencia y la implementación de datos abiertos. La creación del MITIC refuerza el compromiso del gobierno con la modernización tecnológica y la gestión eficiente de la información pública.

### Ley N° 1969/2002 que Regula la información de carácter privado

La Ley N° 1969/2002 modifica y amplía varios artículos de la Ley N° 1682/2001, que regula la información de carácter privado. Esta ley actualiza las disposiciones relativas a la protección de datos personales, asegurando que la información privada sea manejada con el debido respeto a la privacidad y la seguridad de los individuos. Aunque no se centra exclusivamente en datos abiertos, sus disposiciones son cruciales para garantizar que la apertura de datos públicos no comprometa la privacidad de los ciudadanos.

### Ley N° 1682/2001 que Reglamenta la Información de Carácter Privado

La Ley N° 1682/2001 establece las bases para la protección de la información privada en Paraguay. Regula cómo se debe manejar y proteger la información personal, asegurando que los datos privados no sean divulgados sin el consentimiento del titular. Esta ley complementa las iniciativas de datos abiertos, asegurando que la transparencia y el acceso a la información pública no infrinjan los derechos de privacidad de los individuos.

### Relevancia y aplicación

Estas leyes forman un marco integral que apoya la transparencia y la apertura de datos en Paraguay. La Constitución Nacional y la Ley N° 5.282/2014 establecen los principios básicos de acceso a la información pública, mientras que la Ley N° 6.207/2018 proporciona la infraestructura y los mecanismos necesarios para la implementación efectiva de datos abiertos. Las Leyes N° 1969/2002 y N° 1682/2001 aseguran que la protección de datos privados sea una prioridad, equilibrando la necesidad de transparencia con el derecho a la privacidad.

Este marco legal y regulatorio es fundamental para fomentar la participación ciudadana, promover la innovación y asegurar una gestión gubernamental eficiente y responsable en Paraguay.

## 1.3. Principios Fundamentales que sustentan el movimiento de Datos Abiertos


El movimiento de datos abiertos se basa en una serie de principios esenciales que buscan promover la transparencia, la participación ciudadana, la innovación y la eficiencia gubernamental. Estos principios han sido detallados por diversas organizaciones y encuentros internacionales, como el realizado en Sebastopol, California, en 2007, donde se establecieron los "8 principios de Open Data" (Open Knowledge Foundation, 2010). A continuación, se describen los principios clave que sustentan el movimiento de datos abiertos, según Begoña Rivas Rebaque (2018):

### Transparencia

- **Transparencia:** Los datos abiertos permiten una mayor visibilidad de las actividades gubernamentales, aumentando la confianza pública en las instituciones.
- **Rendición de Cuentas:** Facilitan que los ciudadanos y otras entidades puedan evaluar y exigir responsabilidades a las administraciones públicas.

### Participación Ciudadana

- **Empoderamiento Ciudadano:** Los datos abiertos permiten a los ciudadanos participar activamente en la toma de decisiones y en la supervisión de las políticas públicas.
- **Colaboración:** Fomentan la cooperación entre el gobierno y la sociedad civil para resolver problemas públicos y mejorar los servicios.

### Innovación y Desarrollo Económico

- **Fomento de la Innovación:** La disponibilidad de datos abiertos incentiva la creación de nuevas aplicaciones, servicios y productos por parte de emprendedores y desarrolladores.
- **Crecimiento Económico:** Los datos abiertos pueden generar nuevas oportunidades de negocio y fomentar la competitividad al proporcionar igualdad de acceso a la información.

### Eficiencia Gubernamental

- **Mejora en la Gestión:** Permiten una mejor gestión interna de las instituciones gubernamentales al facilitar la interoperabilidad y el intercambio de información.
- **Reducción de Costos:** La reutilización de datos puede reducir los costos asociados con la recopilación y mantenimiento de información.

## 1.4. Comparación del Marco Legal y Regulatorio de Paraguay con otros países


Para comprender mejor la posición de Paraguay en términos de datos abiertos, es útil comparar su marco legal y regulatorio con el de otros países que han avanzado significativamente en esta área. En particular, analizaremos los marcos legales de Estados Unidos, Reino Unido y España, resaltando las características principales y contrastándolas con la situación en Paraguay.

### Estados Unidos

**Características Principales:**

1. **Open Data Policy (E.O. 13642):** Emitida en 2013, establece que los datos del gobierno federal deben ser abiertos y legibles por máquinas por defecto.
2. **Portal Data.gov:** Lanzado en 2009, es una plataforma central que alberga datos abiertos de todas las agencias federales. Se caracteriza por una gran cantidad de datasets registrados. En lo personal, lo considero la mejor fuente gubernamental de información entre los países de occidente, al tratarse de un país con tanta inversión en el manejo de datos e innovación.
3. **Normativas de Estándares:** El gobierno estadounidense ha adoptado normas para garantizar la interoperabilidad y reutilización de los datos abiertos.

**Comparación con Paraguay:**

- **Transparencia:** Al igual que Paraguay, EE.UU. promueve la transparencia, pero lo hace con un enfoque más riguroso en la interoperabilidad de datos.
- **Desarrollo:** Aunque Paraguay esté avanzando con respecto a sus datos abiertos gubernamentales, la diferencia es abismal; el sitio web de datos de EE.UU. está produnfamente desarrollado, con una cantidad de datos mucho más grande, una organización más fuerte y un mejor sistema de consultas.
- **Legislación:** La política de datos abiertos en EE.UU. está respaldada por una orden ejecutiva específica, mientras que Paraguay se apoya en diversas leyes y decretos para promover la apertura de datos.

### Reino Unido

**Características Principales:**

1. **Ley de Transparencia de Información del Sector Público (Freedom of Information Act, FOIA):** Proporciona el derecho a acceder a la información mantenida por las autoridades públicas.
2. **Portal Data.gov.uk:** Centraliza el acceso a los datos abiertos del gobierno, facilitando su reutilización.
3. **Enfoque en Innovación:** Fomenta el uso de datos abiertos para la creación de aplicaciones y servicios innovadores, apoyado por iniciativas como el Data Science Campus de la Oficina Nacional de Estadísticas.

**Comparación con Paraguay:**

- **Transparencia y Participación:** Ambos países promueven la transparencia y la participación ciudadana, aunque el Reino Unido tiene un enfoque más robusto en la promoción de la innovación mediante el uso de datos abiertos.
- **Interoperabilidad y Estándares:** El Reino Unido enfatiza fuertemente la interoperabilidad y la estandarización de datos, áreas donde Paraguay podría mejorar.
- **Educación y Capacitación:** Reino Unido ha implementado programas específicos para capacitar a la población en el uso de datos abiertos, algo que Paraguay podría adoptar para maximizar el impacto de sus iniciativas.

### España

**Características Principales:**

1. **Ley de Transparencia, Acceso a la Información Pública y Buen Gobierno (Ley 19/2013):** Regula el derecho de acceso a la información pública y fomenta la transparencia en la administración pública.
2. **Portal datos.gob.es:** Plataforma que ofrece acceso a datos abiertos de distintas administraciones públicas.
3. **Fomento de la Reutilización:** España promueve la reutilización de los datos del sector público para aplicaciones comerciales y no comerciales, apoyada por iniciativas como el Proyecto Aporta.

**Comparación con Paraguay:**

- **Marco Legal:** España y Paraguay tienen leyes específicas que garantizan el acceso a la información pública. Sin embargo, España tiene un enfoque más estructurado y amplio en términos de fomento de la reutilización de datos.
- **Plataformas de Datos:** Ambos países cuentan con portales de datos abiertos, aunque datos.gob.es ofrece una mayor variedad y cantidad de conjuntos de datos, así como herramientas para su explotación.
- **Educación y Sensibilización:** España ha implementado campañas y programas educativos para sensibilizar a la ciudadanía sobre la importancia y el uso de datos abiertos, una práctica que podría beneficiar a Paraguay.

## 1.5. Desafíos y oportunidades asociados con la implementación de políticas de datos abiertos en Paraguay

La implementación de datos abiertos presenta una serie de desafíos y oportunidades que influyen en la transparencia, la participación ciudadana y la eficiencia gubernamental. En este contexto, es crucial abordar la infraestructura tecnológica, la capacitación, la resistencia al cambio y la protección de datos personales como desafíos clave.

### Desafíos

- **Infraestructura Tecnológica:** La falta de infraestructura tecnológica adecuada puede dificultar la publicación y el acceso a los datos abiertos.

- **Resistencia al Cambio:** Algunas instituciones pueden resistirse a la apertura de datos debido a la cultura organizacional o temor a la rendición de cuentas.
- **Protección de Datos Personales:** Equilibrar la transparencia con la protección de la privacidad es un desafío constante.

### Oportunidades

- **Mejora en la Transparencia Gubernamental:** La implementación de datos abiertos puede aumentar la transparencia y la confianza en las instituciones públicas.
- **Fomento de la Innovación:** Los datos abiertos pueden ser utilizados por emprendedores y desarrolladores para crear nuevos productos y servicios, impulsando la economía.
- **Participación Ciudadana:** Los datos abiertos facilitan la participación de los ciudadanos en la toma de decisiones y el monitoreo de la gestión pública, promoviendo una gobernanza más inclusiva y responsable.
- **Colaboración Interinstitucional:** La apertura de datos puede mejorar la colaboración y la eficiencia entre diferentes entidades gubernamentales, optimizando la gestión de recursos y servicios.

### Distribución de Cada Nivel educativo por Departamento

```python
# Agrupar por departamento y sumar el total de alumnos
top_departamentos = df.groupby('Dpto_descripcion')['TotalAlumnos'].sum().sort_values(ascending=False).head(10).index
df_top_departamentos = df[df['Dpto_descripcion'].isin(top_departamentos)]

# Crear tabla cruzada de niveles educativos vs. descripción de edad
niveles = ['EB1', 'EB2', 'EB3', 'EB4', 'CEGB']
pivot_tables = {}
for nivel in niveles:
    df_top_departamentos[f'{nivel}Total'] = df_top_departamentos[f'{nivel}Varon'] + df_top_departamentos[f'{nivel}Mujer']
    pivot_tables[nivel] = df_top_departamentos.pivot_table(index='Dpto_descripcion', columns='EdadDescripcion', values=f'{nivel}Total', aggfunc='sum')

# Graficar tablas cruzadas
for nivel in niveles:
    plt.figure(figsize=(15, 10))
    sns.heatmap(pivot_tables[nivel], cmap='magma_r', linewidths=.5, annot=True, fmt=".0f")
    plt.xlabel('Edad')
    plt.ylabel('Departamento')
    plt.title(f'Nivel Educativo {nivel} vs. Grupo de Edad en los 10 Principales Departamentos')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

Es esencial aplicar técnicas de análisis rigurosas, validar los resultados y estar alerta a posibles sesgos para garantizar la fiabilidad y precisión de los hallazgos.

## 3.4. Medidas de Mitigación

En esta sección, se proponen medidas específicas para mitigar los riesgos y garantizar un uso ético y responsable de los datos utilizados en el análisis. Estas medidas se basan en estándares de protección de datos, licencias de software y prácticas recomendadas para el manejo seguro de la información.

**Elementos de mitigación y objetivos:**

- **Anonimización de datos sensibles:** Utilizar herramientas de anonimización como el software DataMask para suprimir información personal identificable, como nombres, direcciones y números de identificación, antes de realizar análisis de datos.

- **Acceso controlado:** Implementar un sistema de gestión de acceso basado en roles (RBAC) utilizando software como Keycloak, que permita asignar permisos específicos a usuarios autorizados y registrar y monitorear el acceso a los datos.

- **Capacitación y concienciación:** Brindar capacitación en ética de datos y protección de la privacidad utilizando recursos como los cursos en línea de la Open Data Institute (ODI) y realizar campañas de sensibilización internas sobre buenas prácticas de manejo de datos.

- **Monitoreo y auditoría:** Configurar herramientas de monitoreo y auditoría como Elasticsearch y Kibana para rastrear actividades de acceso y manipulación de datos, detectar anomalías y generar informes de auditoría periódicos.

- **Cumplimiento normativo:** Adherirse a estándares de cumplimiento normativo como el Reglamento General de Protección de Datos (GDPR) de la Unión Europea y obtener licencias de software de código abierto compatibles con GDPR, como Apache License 2.0.

- **Consentimiento informado:** Implementar un proceso de obtención de consentimiento informado utilizando formularios digitales interactivos basados en software como Google Forms o SurveyMonkey, explicando claramente el propósito y alcance del uso de datos.

- **Transparencia:** Publicar una política de privacidad detallada en el sitio web de la institución educativa, que describa cómo se recopilan, almacenan y utilizan los datos de los estudiantes, y proporcionar un mecanismo para que los usuarios ejerzan sus derechos de privacidad.

- **Respaldo y seguridad de datos:** Realizar copias de seguridad regulares de los datos utilizando soluciones como Veeam Backup & Replication, y garantizar la seguridad de los datos en tránsito y en reposo mediante el uso de cifrado SSL/TLS y almacenamiento seguro en servidores protegidos.

La implementación de estas medidas específicas contribuye a garantizar un manejo ético y responsable de los datos, cumpliendo con las normativas legales y protegiendo la privacidad de los individuos. Al aplicar estándares de protección de datos, licencias de software adecuadas y prácticas de seguridad sólidas, se reduce el riesgo de infracciones éticas o legales y se fortalece la confianza en el uso de datos para fines académicos y de formación.

# Conclusión

Este proyecto de investigación y análisis de datos abiertos de Paraguay ha sido una experiencia reveladora en muchos sentidos. A través de la exploración de diversas fuentes de información, desde el portal oficial de datos abiertos del gobierno paraguayo hasta leyes específicas relacionadas con la transparencia y el acceso a la información, se ha podido profundizar en el panorama de los datos abiertos en el país.

Al analizar detenidamente las implicaciones éticas, legales y técnicas asociadas con el uso de estos datos, se ha reafirmado la importancia de promover prácticas de uso responsable y ético de la información pública. Desde la perspectiva legal, la existencia de leyes y licencias que respaldan el acceso y uso de los datos abiertos es fundamental para garantizar la transparencia y la rendición de cuentas en el sector público.

La implementación de medidas para mitigar riesgos y asegurar un uso ético de los datos, como la correcta atribución de la fuente, la actualización de la información y el cumplimiento de las licencias aplicables, son pasos clave hacia una cultura de datos abiertos más robusta y confiable. Además, el uso de herramientas y tecnologías adecuadas, así como la capacitación y concientización de los usuarios, son aspectos cruciales para aprovechar al máximo el potencial de los datos abiertos en beneficio de la sociedad y el desarrollo del país.

Este proyecto subraya la importancia de los datos abiertos como un recurso invaluable para la toma de decisiones informadas, la innovación y el progreso social. Al mismo tiempo, destaca la responsabilidad compartida de todos los actores involucrados en el ecosistema de datos abiertos para garantizar su uso adecuado, ético y responsable en beneficio de la sociedad en su conjunto.
